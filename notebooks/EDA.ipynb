## `capital-gain` column

```python
df['capital-gain'].unique()
```

```python
print(f"Minimum capital gain:- {df['capital-gain'].min()}")
```

```python
print(f"Maximum capital gain: {df['capital-gain'].max()}")
```

```python
list_0_10 = []
list_10_20 = []
list_20_30 = []
list_30_40 = []
list_40_50 = []
list_50_60 = []
list_60_70 = []
list_70_80 = []
list_80_90 = []
list_90_100 = []

for i in df['capital-gain']:
  if i >= 0 and i < 10000:
    list_0_10.append(i)

  elif i >= 10000 and i < 20000:
    list_10_20.append(i)

  elif i >= 20000 and i < 30000:
    list_20_30.append(i)

  elif i >= 30000 and i < 40000:
    list_30_40.append(i)

  elif i >= 40000 and i < 50000:
    list_40_50.append(i)

  elif i >= 50000 and i < 60000:
    list_50_60.append(i)

  elif i >= 60000 and i < 70000:
    list_60_70.append(i)

  elif i >= 70000 and i < 80000:
    list_70_80.append(i)

  elif i >= 80000 and i < 90000:
    list_80_90.append(i)

  elif i >= 90000 and i < 100000:
    list_90_100.append(i)

```

```python
X_gain = ['0-10000',
     '10000-20000',
     '20000-30000',
     '30000-40000',
     '40000-50000',
     '50000-60000',
     '60000-70000',
     '70000-80000',
     '80000-90000',
     '90000-100000'
]


Y_gain = [len(list_0_10),
    len(list_10_20),
    len(list_20_30),
    len(list_30_40),
    len(list_40_50),
    len(list_50_60),
    len(list_60_70),
    len(list_70_80),
    len(list_80_90),
    len(list_90_100)
]
```

```python
plt.style.use('ggplot')

plt.figure(figsize = (10, 9))

sn.barplot(x = X_gain, y = Y_gain,
           palette = 'YlGnBu',
           width=0.4)


for i in range(len(X_gain)):
  plt.text(i, Y_gain[i], Y_gain[i], ha = 'center', va = 'bottom')

plt.title('Capital gain')
plt.xlabel('Net gain')
plt.xticks(rotation = 45)

plt.ylabel('Count of person in the bracket')

plt.show()
```

While respondents have achieved a diverse range of capital gains, a significant **`majority`** still falls within the **`0-10000`** brackets. Notably, only **`159 individuals`** have experienced an impressive capital gain, ranging between **`90000-100000`**.

The workforce exhibits a diverse range of working hours, with a minimum commitment of 1 hour per week and a maximum limit of 99 hours. The majority of individuals seem to be engaged in service-based jobs, as evidenced by a concentration within the 40-50 hours per week bracket.

Interestingly, there is a notable group working exceptionally long hours, with a minority dedicating almost 90-100 hours per week. Additionally, responses show a relatively balanced distribution between the 30-40 and 50-60 hours per week categories.

This spectrum of working hours suggests a varied professional landscape, where some individuals opt for part-time or flexible arrangements, while others are deeply immersed in demanding roles requiring significant time commitments.


**Job Categorization:**

1. **White-Collar Jobs:**
   - *Exec-managerial:* Positions involving executive or managerial responsibilities.
   - *Prof-specialty:* Professional and specialized roles requiring specific expertise.
   - *Tech-support:* Jobs providing technical support and assistance.

2. **Service Jobs:**
   - *Adm-clerical:* Administrative and clerical roles.
   - *Other-service:* Various service-oriented positions not specified elsewhere.
   - *Priv-house-serv:* Private household services and domestic roles.

3. **Sales Jobs:**
   - *Sales:* Occupations focused on sales and revenue generation.

4. **Blue-Collar Jobs:**
   - *Handlers-cleaners:* Roles related to handling and cleaning tasks.
   - *Craft-repair:* Positions involving craftsmanship and repair work.
   - *Transport-moving:* Jobs associated with transportation and moving goods.
   - *Farming-fishing:* Agricultural and fishing-related occupations.
   - *Machine-op-inspct:* Roles centered around machine operation and inspection.

5. **Protective Services:**
   - *Protective-serv:* Jobs in protective services ensuring safety and security.

6. **Armed Forces:**
   - *Armed-Forces:* Military occupations serving in the armed forces.

This categorization aims to provide a clearer understanding of different job types within distinct sectors.
