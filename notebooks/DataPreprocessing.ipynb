### Applying Ordinal Encoder

```python
order = [['Preschool', '1st-4th', '5th-6th', '7th-8th', '9th', '10th', '11th', '12th', 'HS-grad', 'Some-college', 'Assoc-acdm', 'Assoc-voc', 'Bachelors', 'Masters', 'Doctorate', 'Prof-school']]
```

```python
ord_enc = OrdinalEncoder(categories = order)
```

```python
ord_enc.fit(df2[['new_education']])
```

```python
df2['new_education'] = ord_enc.transform(df2[['new_education']])
```

## Preprocessing `new_workclass` column

### Applying `OneHotEncoder`

```python
ohe = OneHotEncoder(drop = 'first', sparse_output = False, dtype = np.float32)
```

```python
ohe.fit(df[['new_workclass', 'new_occupation']])
```

```python
Encoded_values = ohe.transform(df[['new_workclass', 'new_occupation']])
```

```python
Encoded_values
```

```python
feature_names = ohe.get_feature_names_out(input_features = ['new_workclass', 'new_occupation'])
```

```python
feature_names
```

```python
new_encoded_dataframe = pd.DataFrame(Encoded_values, columns = feature_names)
```

### Forming New Dataframe with encoded columns

```python
df4 = pd.concat([df2, new_encoded_dataframe], axis = 'columns')
```

The `marriage` column in the dataset appears to contribute little to the predictive accuracy of whether an individual's salary will exceed 50k or not. Consequently, it is deemed dispensable, along with the outdated `occupation` and `workplace` columns. Removal of these columns aims to enhance the model's efficiency and streamline the dataset by focusing on the most relevant features for the salary prediction task.

### Dropping older `workclass`, `occupation`, `marital_status` columns

```python
df5 = df4.drop(['new_workclass', 'new_occupation', 'new_marital_status'], axis = 'columns')
```
